````python
feature=['price','minimum_nights','number_of_reviews', 'reviews_per_month','calculated_host_listings_count','availability_365']
fig,axs= plt.subplots(nrows=2,ncols=3,figsize=(15,10))

# flatten the axs array to iterate over it with a single loop
axs=axs.flatten()
for i,col in enumerate(feature):
  sns.displot(airbnb[col],ax=axs[i])
  axs[i].axvline(airbnb[col].mean(),color='magenta',linestyle='dashed',linewidth=2)
  axs[i].axvline(airbnb[col].median(),color='cyan',linestyle='dashed',linewidth=2)
  axs[i].set_title(col)
  plt.tight_layout()
  plt.show()
````
